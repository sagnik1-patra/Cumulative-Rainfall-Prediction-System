```python
import os
import json
import random
import warnings
from datetime import datetime

import joblib
import numpy as np
import pandas as pd
import yaml

from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.models import Sequential


# ================================================================
# 1. GENERAL SETTINGS
# ================================================================

warnings.filterwarnings("ignore")

RANDOM_STATE = 42

random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

PROJECT_FOLDER = (
    r"C:\Users\NXTWAVE\Downloads"
    r"\Cumulative Rainfall Prediction System"
)

CSV_PATH = os.path.join(
    PROJECT_FOLDER,
    "daily_rainfall_ndb.csv"
)

H5_MODEL_PATH = os.path.join(
    PROJECT_FOLDER,
    "cumulative_rainfall_ann_model.h5"
)

PKL_MODEL_PATH = os.path.join(
    PROJECT_FOLDER,
    "cumulative_rainfall_model.pkl"
)

JSON_PATH = os.path.join(
    PROJECT_FOLDER,
    "cumulative_rainfall_config.json"
)

YAML_PATH = os.path.join(
    PROJECT_FOLDER,
    "cumulative_rainfall_config.yaml"
)

RESULT_CSV_PATH = os.path.join(
    PROJECT_FOLDER,
    "cumulative_rainfall_results.csv"
)

TARGET_COLUMN = "Progressive (m.m)"
DAILY_RAINFALL_COLUMN = "Daily Rainfall (m.m)"
DATE_COLUMN = "Date"
DISTRICT_COLUMN = "District"


# ================================================================
# 2. HELPER FUNCTIONS
# ================================================================

def convert_to_python_type(value):
    """
    Converts NumPy data types into standard Python data types.
    This prevents JSON and YAML serialization errors.
    """

    if isinstance(value, (np.integer,)):
        return int(value)

    if isinstance(value, (np.floating,)):
        return float(value)

    if isinstance(value, np.ndarray):
        return value.tolist()

    return value


def calculate_metrics(actual, predicted):
    """
    Calculates regression evaluation metrics.
    """

    actual = np.asarray(actual).reshape(-1)
    predicted = np.asarray(predicted).reshape(-1)

    mae = mean_absolute_error(actual, predicted)
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)

    # R² needs at least two observations.
    if len(actual) >= 2:
        r2 = r2_score(actual, predicted)
    else:
        r2 = 0.0

    # Avoid division by zero while calculating percentage error.
    non_zero_mask = actual != 0

    if np.any(non_zero_mask):
        mape = np.mean(
            np.abs(
                (
                    actual[non_zero_mask] -
                    predicted[non_zero_mask]
                ) / actual[non_zero_mask]
            )
        ) * 100
    else:
        mape = 0.0

    return {
        "mae": float(mae),
        "mse": float(mse),
        "rmse": float(rmse),
        "r2_score": float(r2),
        "mape_percentage": float(mape)
    }


def create_ann_model(number_of_features):
    """
    Builds and compiles an Artificial Neural Network regression model.
    """

    model = Sequential([
        Input(shape=(number_of_features,)),

        Dense(
            64,
            activation="relu"
        ),

        Dropout(0.15),

        Dense(
            32,
            activation="relu"
        ),

        Dropout(0.10),

        Dense(
            16,
            activation="relu"
        ),

        Dense(
            1,
            activation="linear"
        )
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(
            learning_rate=0.001
        ),
        loss="mean_squared_error",
        metrics=["mean_absolute_error"]
    )

    return model


# ================================================================
# 3. CREATE AND VERIFY PROJECT FOLDER
# ================================================================

os.makedirs(
    PROJECT_FOLDER,
    exist_ok=True
)

if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(
        "\nDataset was not found.\n"
        f"Expected location:\n{CSV_PATH}\n\n"
        "Check the folder name and CSV filename."
    )


# ================================================================
# 4. LOAD DATASET
# ================================================================

print("=" * 70)
print("CUMULATIVE RAINFALL PREDICTION SYSTEM")
print("=" * 70)

print("\nLoading dataset...")

data = pd.read_csv(CSV_PATH)

print("Dataset loaded successfully.")
print(f"Dataset shape: {data.shape}")

print("\nOriginal columns:")
for column in data.columns:
    print(f" - {column}")

print("\nFirst five records:")
print(data.head())


# ================================================================
# 5. VALIDATE REQUIRED COLUMNS
# ================================================================

required_columns = [
    DISTRICT_COLUMN,
    DATE_COLUMN,
    DAILY_RAINFALL_COLUMN,
    TARGET_COLUMN
]

missing_columns = [
    column
    for column in required_columns
    if column not in data.columns
]

if missing_columns:
    raise ValueError(
        "The following required columns are missing: "
        + ", ".join(missing_columns)
    )


# ================================================================
# 6. CLEAN THE DATASET
# ================================================================

# Remove extra spaces from column names.
data.columns = data.columns.str.strip()

# Remove completely empty rows.
data.dropna(
    how="all",
    inplace=True
)

# Clean district names.
data[DISTRICT_COLUMN] = (
    data[DISTRICT_COLUMN]
    .astype(str)
    .str.strip()
)

# Convert the Date column.
# The dataset uses dates like 01-09-2025, so dayfirst=True is used.
data[DATE_COLUMN] = pd.to_datetime(
    data[DATE_COLUMN],
    dayfirst=True,
    errors="coerce"
)

# Convert rainfall columns to numeric values.
data[DAILY_RAINFALL_COLUMN] = pd.to_numeric(
    data[DAILY_RAINFALL_COLUMN],
    errors="coerce"
)

data[TARGET_COLUMN] = pd.to_numeric(
    data[TARGET_COLUMN],
    errors="coerce"
)

# Remove rows where the date or target is unavailable.
data.dropna(
    subset=[
        DATE_COLUMN,
        TARGET_COLUMN
    ],
    inplace=True
)

# Fill missing daily rainfall with the median.
daily_rainfall_median = data[
    DAILY_RAINFALL_COLUMN
].median()

if pd.isna(daily_rainfall_median):
    daily_rainfall_median = 0.0

data[DAILY_RAINFALL_COLUMN] = (
    data[DAILY_RAINFALL_COLUMN]
    .fillna(daily_rainfall_median)
)

# Negative rainfall values are invalid.
data[DAILY_RAINFALL_COLUMN] = (
    data[DAILY_RAINFALL_COLUMN]
    .clip(lower=0)
)

data[TARGET_COLUMN] = (
    data[TARGET_COLUMN]
    .clip(lower=0)
)

# Sort records chronologically.
data.sort_values(
    by=[
        DISTRICT_COLUMN,
        DATE_COLUMN
    ],
    inplace=True
)

data.reset_index(
    drop=True,
    inplace=True
)

print("\nDataset cleaned successfully.")
print(f"Cleaned dataset shape: {data.shape}")


# ================================================================
# 7. FEATURE ENGINEERING
# ================================================================

# Basic date-related features.
data["Year"] = data[DATE_COLUMN].dt.year
data["Month"] = data[DATE_COLUMN].dt.month
data["Day"] = data[DATE_COLUMN].dt.day
data["DayOfYear"] = data[DATE_COLUMN].dt.dayofyear
data["DayOfWeek"] = data[DATE_COLUMN].dt.dayofweek
data["WeekOfYear"] = (
    data[DATE_COLUMN]
    .dt.isocalendar()
    .week
    .astype(int)
)

data["Quarter"] = data[DATE_COLUMN].dt.quarter

data["IsWeekend"] = (
    data["DayOfWeek"]
    .isin([5, 6])
    .astype(int)
)

# Cyclical date features help models understand that months and days repeat.
data["MonthSin"] = np.sin(
    2 * np.pi * data["Month"] / 12
)

data["MonthCos"] = np.cos(
    2 * np.pi * data["Month"] / 12
)

data["DayOfYearSin"] = np.sin(
    2 * np.pi * data["DayOfYear"] / 365.25
)

data["DayOfYearCos"] = np.cos(
    2 * np.pi * data["DayOfYear"] / 365.25
)

# Lag rainfall features.
# The shift is performed separately for each district.
data["PreviousDayRainfall"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .shift(1)
)

data["RainfallLag2"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .shift(2)
)

data["RainfallLag3"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .shift(3)
)

# Previous progressive rainfall values.
data["PreviousProgressiveRainfall"] = (
    data
    .groupby(DISTRICT_COLUMN)[TARGET_COLUMN]
    .shift(1)
)

data["ProgressiveLag2"] = (
    data
    .groupby(DISTRICT_COLUMN)[TARGET_COLUMN]
    .shift(2)
)

# Rolling rainfall averages.
# shift(1) prevents the current target row from leaking into historical data.
data["RainfallRollingMean3"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .transform(
        lambda values:
        values.shift(1).rolling(
            window=3,
            min_periods=1
        ).mean()
    )
)

data["RainfallRollingSum3"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .transform(
        lambda values:
        values.shift(1).rolling(
            window=3,
            min_periods=1
        ).sum()
    )
)

data["RainfallRollingMean7"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .transform(
        lambda values:
        values.shift(1).rolling(
            window=7,
            min_periods=1
        ).mean()
    )
)

data["RainfallRollingSum7"] = (
    data
    .groupby(DISTRICT_COLUMN)[DAILY_RAINFALL_COLUMN]
    .transform(
        lambda values:
        values.shift(1).rolling(
            window=7,
            min_periods=1
        ).sum()
    )
)

# Fill initial lag and rolling values.
lag_columns = [
    "PreviousDayRainfall",
    "RainfallLag2",
    "RainfallLag3",
    "PreviousProgressiveRainfall",
    "ProgressiveLag2",
    "RainfallRollingMean3",
    "RainfallRollingSum3",
    "RainfallRollingMean7",
    "RainfallRollingSum7"
]

for column in lag_columns:
    column_median = data[column].median()

    if pd.isna(column_median):
        column_median = 0.0

    data[column] = data[column].fillna(column_median)

print("Feature engineering completed.")


# ================================================================
# 8. SELECT INPUT AND TARGET COLUMNS
# ================================================================

numeric_features = [
    DAILY_RAINFALL_COLUMN,
    "Year",
    "Month",
    "Day",
    "DayOfYear",
    "DayOfWeek",
    "WeekOfYear",
    "Quarter",
    "IsWeekend",
    "MonthSin",
    "MonthCos",
    "DayOfYearSin",
    "DayOfYearCos",
    "PreviousDayRainfall",
    "RainfallLag2",
    "RainfallLag3",
    "PreviousProgressiveRainfall",
    "ProgressiveLag2",
    "RainfallRollingMean3",
    "RainfallRollingSum3",
    "RainfallRollingMean7",
    "RainfallRollingSum7"
]

categorical_features = [
    DISTRICT_COLUMN
]

feature_columns = (
    categorical_features +
    numeric_features
)

X = data[feature_columns].copy()
y = data[TARGET_COLUMN].copy()


# ================================================================
# 9. CHRONOLOGICAL TRAIN-TEST SPLIT
# ================================================================

number_of_rows = len(data)

if number_of_rows < 5:
    raise ValueError(
        "At least five valid rows are needed to train and test the models."
    )

# Use approximately 80% data for training.
split_index = int(number_of_rows * 0.80)

# Keep at least two test records when possible.
if number_of_rows - split_index < 2:
    split_index = number_of_rows - 2

# Keep at least three training records.
split_index = max(split_index, 3)

X_train = X.iloc[:split_index].copy()
X_test = X.iloc[split_index:].copy()

y_train = y.iloc[:split_index].copy()
y_test = y.iloc[split_index:].copy()

train_dates = data[DATE_COLUMN].iloc[:split_index].copy()
test_dates = data[DATE_COLUMN].iloc[split_index:].copy()

print("\nChronological data split completed.")
print(f"Training records: {len(X_train)}")
print(f"Testing records : {len(X_test)}")


# ================================================================
# 10. CREATE PREPROCESSING PIPELINE
# ================================================================

numeric_pipeline = Pipeline(
    steps=[
        (
            "imputer",
            SimpleImputer(
                strategy="median"
            )
        ),
        (
            "scaler",
            StandardScaler()
        )
    ]
)

# Compatibility handling for different scikit-learn versions.
try:
    categorical_encoder = OneHotEncoder(
        handle_unknown="ignore",
        sparse_output=False
    )
except TypeError:
    categorical_encoder = OneHotEncoder(
        handle_unknown="ignore",
        sparse=False
    )

categorical_pipeline = Pipeline(
    steps=[
        (
            "imputer",
            SimpleImputer(
                strategy="most_frequent"
            )
        ),
        (
            "encoder",
            categorical_encoder
        )
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        (
            "numeric",
            numeric_pipeline,
            numeric_features
        ),
        (
            "categorical",
            categorical_pipeline,
            categorical_features
        )
    ],
    remainder="drop"
)


# ================================================================
# 11. PREPROCESS TRAINING AND TEST DATA
# ================================================================

X_train_processed = preprocessor.fit_transform(
    X_train
)

X_test_processed = preprocessor.transform(
    X_test
)

X_all_processed = preprocessor.transform(
    X
)

# TensorFlow works well with float32 values.
X_train_processed = np.asarray(
    X_train_processed
).astype(np.float32)

X_test_processed = np.asarray(
    X_test_processed
).astype(np.float32)

X_all_processed = np.asarray(
    X_all_processed
).astype(np.float32)

print(
    "\nNumber of processed features:",
    X_train_processed.shape[1]
)


# ================================================================
# 12. TRAIN RANDOM FOREST MODEL
# ================================================================

print("\nTraining Random Forest model...")

random_forest_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

random_forest_model.fit(
    X_train_processed,
    y_train
)

rf_train_predictions = random_forest_model.predict(
    X_train_processed
)

rf_test_predictions = random_forest_model.predict(
    X_test_processed
)

rf_all_predictions = random_forest_model.predict(
    X_all_processed
)

rf_train_metrics = calculate_metrics(
    y_train,
    rf_train_predictions
)

rf_test_metrics = calculate_metrics(
    y_test,
    rf_test_predictions
)

print("Random Forest training completed.")


# ================================================================
# 13. SCALE TARGET FOR ANN
# ================================================================

target_scaler = StandardScaler()

y_train_scaled = target_scaler.fit_transform(
    y_train.to_numpy().reshape(-1, 1)
).astype(np.float32)

y_test_scaled = target_scaler.transform(
    y_test.to_numpy().reshape(-1, 1)
).astype(np.float32)


# ================================================================
# 14. BUILD AND TRAIN ANN MODEL
# ================================================================

print("\nTraining Artificial Neural Network...")

ann_model = create_ann_model(
    X_train_processed.shape[1]
)

early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=50,
    restore_best_weights=True,
    verbose=0
)

# For very small datasets, validation_split can leave too few records.
if len(X_train_processed) >= 10:
    validation_split = 0.20
else:
    validation_split = 0.0

callbacks = []

if validation_split > 0:
    callbacks.append(early_stopping)

history = ann_model.fit(
    X_train_processed,
    y_train_scaled,
    epochs=500,
    batch_size=min(8, len(X_train_processed)),
    validation_split=validation_split,
    callbacks=callbacks,
    verbose=0,
    shuffle=False
)

ann_train_predictions_scaled = ann_model.predict(
    X_train_processed,
    verbose=0
)

ann_test_predictions_scaled = ann_model.predict(
    X_test_processed,
    verbose=0
)

ann_all_predictions_scaled = ann_model.predict(
    X_all_processed,
    verbose=0
)

ann_train_predictions = (
    target_scaler
    .inverse_transform(
        ann_train_predictions_scaled
    )
    .reshape(-1)
)

ann_test_predictions = (
    target_scaler
    .inverse_transform(
        ann_test_predictions_scaled
    )
    .reshape(-1)
)

ann_all_predictions = (
    target_scaler
    .inverse_transform(
        ann_all_predictions_scaled
    )
    .reshape(-1)
)

ann_train_metrics = calculate_metrics(
    y_train,
    ann_train_predictions
)

ann_test_metrics = calculate_metrics(
    y_test,
    ann_test_predictions
)

print("ANN training completed.")


# ================================================================
# 15. CHOOSE THE BETTER MODEL
# ================================================================

if ann_test_metrics["rmse"] < rf_test_metrics["rmse"]:
    best_model_name = "Artificial Neural Network"
    best_test_predictions = ann_test_predictions
    best_all_predictions = ann_all_predictions
    best_test_metrics = ann_test_metrics
else:
    best_model_name = "Random Forest Regressor"
    best_test_predictions = rf_test_predictions
    best_all_predictions = rf_all_predictions
    best_test_metrics = rf_test_metrics

print("\nBest model based on test RMSE:")
print(best_model_name)


# ================================================================
# 16. SAVE H5 MODEL
# ================================================================

ann_model.save(
    H5_MODEL_PATH,
    include_optimizer=True
)

print(f"\nH5 model saved:\n{H5_MODEL_PATH}")


# ================================================================
# 17. SAVE PKL MODEL PACKAGE
# ================================================================

# The PKL file contains everything required for preprocessing
# and Random Forest prediction.
model_package = {
    "project_name": "Cumulative Rainfall Prediction System",
    "model_name": "Random Forest Regressor",
    "preprocessor": preprocessor,
    "random_forest_model": random_forest_model,
    "target_scaler": target_scaler,
    "feature_columns": feature_columns,
    "numeric_features": numeric_features,
    "categorical_features": categorical_features,
    "target_column": TARGET_COLUMN,
    "date_column": DATE_COLUMN,
    "district_column": DISTRICT_COLUMN,
    "daily_rainfall_column": DAILY_RAINFALL_COLUMN,
    "random_state": RANDOM_STATE,
    "created_at": datetime.now().isoformat()
}

joblib.dump(
    model_package,
    PKL_MODEL_PATH
)

print(f"\nPKL model package saved:\n{PKL_MODEL_PATH}")


# ================================================================
# 18. CREATE AND SAVE RESULT CSV
# ================================================================

result_data = data[
    [
        DISTRICT_COLUMN,
        DATE_COLUMN,
        DAILY_RAINFALL_COLUMN,
        TARGET_COLUMN
    ]
].copy()

result_data.rename(
    columns={
        TARGET_COLUMN: "Actual Progressive Rainfall (m.m)"
    },
    inplace=True
)

result_data["Random Forest Prediction (m.m)"] = (
    rf_all_predictions
)

result_data["ANN Prediction (m.m)"] = (
    ann_all_predictions
)

result_data["Best Model Prediction (m.m)"] = (
    best_all_predictions
)

result_data["Random Forest Error"] = (
    result_data["Actual Progressive Rainfall (m.m)"] -
    result_data["Random Forest Prediction (m.m)"]
)

result_data["ANN Error"] = (
    result_data["Actual Progressive Rainfall (m.m)"] -
    result_data["ANN Prediction (m.m)"]
)

result_data["Data Split"] = [
    "Training" if index < split_index else "Testing"
    for index in range(len(result_data))
]

# Make the date easy to read in CSV.
result_data[DATE_COLUMN] = (
    result_data[DATE_COLUMN]
    .dt.strftime("%d-%m-%Y")
)

numeric_result_columns = [
    "Actual Progressive Rainfall (m.m)",
    "Random Forest Prediction (m.m)",
    "ANN Prediction (m.m)",
    "Best Model Prediction (m.m)",
    "Random Forest Error",
    "ANN Error"
]

result_data[numeric_result_columns] = (
    result_data[numeric_result_columns]
    .round(4)
)

result_data.to_csv(
    RESULT_CSV_PATH,
    index=False
)

print(f"\nResult CSV saved:\n{RESULT_CSV_PATH}")


# ================================================================
# 19. PREPARE JSON AND YAML INFORMATION
# ================================================================

configuration = {
    "project": {
        "name": "Cumulative Rainfall Prediction System",
        "description": (
            "Machine learning system for predicting progressive "
            "cumulative rainfall from daily and historical rainfall data."
        ),
        "created_at": datetime.now().isoformat(),
        "random_state": RANDOM_STATE
    },

    "dataset": {
        "file_path": CSV_PATH,
        "original_rows": int(number_of_rows),
        "training_rows": int(len(X_train)),
        "testing_rows": int(len(X_test)),
        "districts": sorted(
            data[DISTRICT_COLUMN]
            .dropna()
            .astype(str)
            .unique()
            .tolist()
        ),
        "start_date": data[DATE_COLUMN].min().strftime(
            "%Y-%m-%d"
        ),
        "end_date": data[DATE_COLUMN].max().strftime(
            "%Y-%m-%d"
        ),
        "target_column": TARGET_COLUMN
    },

    "features": {
        "categorical_features": categorical_features,
        "numeric_features": numeric_features,
        "total_raw_features": int(len(feature_columns)),
        "processed_feature_count": int(
            X_train_processed.shape[1]
        )
    },

    "models": {
        "random_forest": {
            "algorithm": "RandomForestRegressor",
            "parameters": {
                "n_estimators": 300,
                "max_depth": 8,
                "min_samples_split": 2,
                "min_samples_leaf": 1,
                "random_state": RANDOM_STATE
            },
            "training_metrics": rf_train_metrics,
            "testing_metrics": rf_test_metrics
        },

        "artificial_neural_network": {
            "algorithm": "Keras Sequential ANN",
            "layers": [
                {
                    "units": 64,
                    "activation": "relu"
                },
                {
                    "layer": "Dropout",
                    "rate": 0.15
                },
                {
                    "units": 32,
                    "activation": "relu"
                },
                {
                    "layer": "Dropout",
                    "rate": 0.10
                },
                {
                    "units": 16,
                    "activation": "relu"
                },
                {
                    "units": 1,
                    "activation": "linear"
                }
            ],
            "epochs_requested": 500,
            "epochs_completed": int(
                len(history.history["loss"])
            ),
            "batch_size": int(
                min(8, len(X_train_processed))
            ),
            "validation_split": float(
                validation_split
            ),
            "training_metrics": ann_train_metrics,
            "testing_metrics": ann_test_metrics
        }
    },

    "best_model": {
        "name": best_model_name,
        "selection_metric": "Lowest test RMSE",
        "test_metrics": best_test_metrics
    },

    "saved_files": {
        "h5_model": H5_MODEL_PATH,
        "pkl_model": PKL_MODEL_PATH,
        "json_configuration": JSON_PATH,
        "yaml_configuration": YAML_PATH,
        "result_csv": RESULT_CSV_PATH
    },

    "important_note": (
        "The available dataset is very small. Predictions and evaluation "
        "metrics should be treated as experimental until more historical "
        "rainfall records are added."
    )
}


# ================================================================
# 20. SAVE JSON
# ================================================================

with open(
    JSON_PATH,
    "w",
    encoding="utf-8"
) as json_file:
    json.dump(
        configuration,
        json_file,
        indent=4,
        ensure_ascii=False,
        default=convert_to_python_type
    )

print(f"\nJSON file saved:\n{JSON_PATH}")


# ================================================================
# 21. SAVE YAML
# ================================================================

with open(
    YAML_PATH,
    "w",
    encoding="utf-8"
) as yaml_file:
    yaml.safe_dump(
        configuration,
        yaml_file,
        sort_keys=False,
        allow_unicode=True
    )

print(f"\nYAML file saved:\n{YAML_PATH}")


# ================================================================
# 22. DISPLAY MODEL RESULTS
# ================================================================

print("\n" + "=" * 70)
print("RANDOM FOREST RESULTS")
print("=" * 70)

print("\nTraining metrics:")
for metric_name, metric_value in rf_train_metrics.items():
    print(
        f"{metric_name}: "
        f"{metric_value:.6f}"
    )

print("\nTesting metrics:")
for metric_name, metric_value in rf_test_metrics.items():
    print(
        f"{metric_name}: "
        f"{metric_value:.6f}"
    )


print("\n" + "=" * 70)
print("ARTIFICIAL NEURAL NETWORK RESULTS")
print("=" * 70)

print("\nTraining metrics:")
for metric_name, metric_value in ann_train_metrics.items():
    print(
        f"{metric_name}: "
        f"{metric_value:.6f}"
    )

print("\nTesting metrics:")
for metric_name, metric_value in ann_test_metrics.items():
    print(
        f"{metric_name}: "
        f"{metric_value:.6f}"
    )


# ================================================================
# 23. DISPLAY TEST PREDICTIONS
# ================================================================

test_result = pd.DataFrame({
    "Date": test_dates.dt.strftime("%d-%m-%Y").values,
    "Actual Progressive Rainfall": y_test.values,
    "Random Forest Prediction": rf_test_predictions,
    "ANN Prediction": ann_test_predictions,
    "Best Model Prediction": best_test_predictions
})

test_result[
    [
        "Actual Progressive Rainfall",
        "Random Forest Prediction",
        "ANN Prediction",
        "Best Model Prediction"
    ]
] = test_result[
    [
        "Actual Progressive Rainfall",
        "Random Forest Prediction",
        "ANN Prediction",
        "Best Model Prediction"
    ]
].round(4)

print("\n" + "=" * 70)
print("TEST SET PREDICTIONS")
print("=" * 70)

print(test_result.to_string(index=False))


# ================================================================
# 24. FINAL SUMMARY
# ================================================================

print("\n" + "=" * 70)
print("PROCESS COMPLETED SUCCESSFULLY")
print("=" * 70)

print(f"\nBest model: {best_model_name}")
print(
    "Best model test RMSE:",
    round(
        best_test_metrics["rmse"],
        6
    )
)

print("\nGenerated files:")

generated_files = [
    H5_MODEL_PATH,
    PKL_MODEL_PATH,
    JSON_PATH,
    YAML_PATH,
    RESULT_CSV_PATH
]

for file_path in generated_files:
    if os.path.exists(file_path):
        file_size = os.path.getsize(file_path)

        print(
            f"[SAVED] {file_path} "
            f"({file_size} bytes)"
        )
    else:
        print(
            f"[NOT FOUND] {file_path}"
        )

print("\nAll files were saved inside:")
print(PROJECT_FOLDER)
```
